```python
rr_data = rr_pd.read_csv('/content/drive/MyDrive/Datasets/preprocessed_rolls_royce.csv')
```

```python
rr_data
```

```python
rr_data.shape
```

```python
import warnings as rr_w
rr_w.filterwarnings('ignore')
import numpy as rr_nmp
```

```python
def rolls_royce_stock_db(S, N):
  rstock_X = []
  rstock_Y = []
  for f in range(len(S)):
    rest = f + N
    if rest > len(S)-1:
      break
    rest_ip, rest_op = S[f:rest], S[rest]
    rstock_X.append(rest_ip)
    rstock_Y.append(rest_op)
  return rr_nmp.array(rstock_X), rr_nmp.array(rstock_Y)
```

```python
# Training data
N = 10
rstock_Xt, rstock_Yt = rolls_royce_stock_db(rr_data['Close'][:-100].tolist(), N)

# Testing data
rstock_Xs, rstock_Ys = rolls_royce_stock_db(rr_data['Close'][-100:].tolist(), N)
for m in range(len(rstock_Xs)):
  print(rstock_Xs[m], rstock_Ys[m])
```

```python
rstock_Xt.shape
```

```python
rstock_Xs.shape
```

```python
# Reshaping training data
rstock_Xt = rstock_Xt.reshape(rstock_Xt.shape[0],rstock_Xt.shape[1] , 1)
print("Training Data after Reshaped : ", rstock_Xt.shape)
```

```python
# reshaping testing data
rstock_Xs = rstock_Xs.reshape(rstock_Xs.shape[0],rstock_Xs.shape[1] , 1)
print("Testing Data after Reshaped : ", rstock_Xs.shape)
```

```python
from tensorflow.keras.layers import LSTM as rstock_longstm
from tensorflow.keras.layers import SimpleRNN as rstock_srnn
from tensorflow.keras.models import Sequential as rstock_seq
from tensorflow.keras.layers import Dense as rstock_dnse
from sklearn.metrics import mean_absolute_error as rstock_mae
from sklearn.metrics import mean_squared_error as rstock_mse
from sklearn.metrics import r2_score as rstock_r2sc
from tensorflow.keras.layers import Bidirectional as rstock_lonbdr
import math as rstock_math

rstock_actvn = 'relu'
rstock_loss = 'mean_squared_error'
```

```python
rstock_Xt = rstock_Xt.astype(rr_nmp.float32)
rstock_Xs = rstock_Xs.astype(rr_nmp.float32)
print(rstock_Xt.shape)
```

**LSTM model Implementation (Optimisers)**

```python
for rstock_op in ['adam', 'rmsprop', 'adamax']:
  print("...............................................................................")
  print("              LSTM using the Optimizer ", rstock_op)
  print("...............................................................................\n")
  rstock_mod = rstock_seq()
  rstock_mod.add(rstock_longstm(100, input_shape=(None,1), activation=rstock_actvn))
  rstock_mod.add(rstock_dnse(1))

  rstock_mod.compile(loss= rstock_loss, optimizer= rstock_op)
  rstock_hstry = rstock_mod.fit(rstock_Xt, rstock_Yt, epochs = 10, batch_size = 200)
  rstock_prd = rstock_mod.predict(rstock_Xs)

  print("\nEvaluation of Rolls Royce on LSTM\n")

  print("Score of R2   :", rstock_r2sc(rstock_Ys, rstock_prd)*100)
  print("Score of MAE  :", rstock_mae(rstock_Ys, rstock_prd))
  print("Score of MSE  :", rstock_mse(rstock_Ys, rstock_prd))
  print("Score of RMSE :", rstock_math.sqrt(rstock_mse(rstock_Ys, rstock_prd)))
```

**LSTM model Implementation (epochs)**

```python
for rstock_epo in [20, 30, 50]:
  print("...............................................................................")
  print("              LSTM using the Epochs ", rstock_epo)
  print("...............................................................................\n")
  rstock_mod = rstock_seq()
  rstock_mod.add(rstock_longstm(100, input_shape=(None,1), activation=rstock_actvn))
  rstock_mod.add(rstock_dnse(1))

  rstock_mod.compile(loss= rstock_loss, optimizer= 'rmsprop')
  rstock_hstry = rstock_mod.fit(rstock_Xt, rstock_Yt, epochs = rstock_epo, batch_size = 200)
  rstock_prd = rstock_mod.predict(rstock_Xs)

  print("\nEvaluation of Rolls Royce Stock on LSTM\n")

  print("Score of R2   :", rstock_r2sc(rstock_Ys, rstock_prd)*100)
  print("Score of MAE  :", rstock_mae(rstock_Ys, rstock_prd))
  print("Score of MSE  :", rstock_mse(rstock_Ys, rstock_prd))
  print("Score of RMSE :", rstock_math.sqrt(rstock_mse(rstock_Ys, rstock_prd)))
```

**Stacked LSTM (Optimizers)**

```python
for rstock_op in ['adam', 'rmsprop', 'adamax']:
  print("...............................................................................")
  print("             Stacked LSTM using the Optimizer ", rstock_op)
  print("...............................................................................\n")
  rstock_mod = rstock_seq()
  rstock_mod.add(rstock_longstm(200, input_shape=(None,1), activation=rstock_actvn, return_sequences=True))
  rstock_mod.add(rstock_longstm(100, input_shape=(None,1), activation=rstock_actvn))
  rstock_mod.add(rstock_dnse(1))

  rstock_mod.compile(loss= rstock_loss, optimizer= rstock_op)
  rstock_hstry = rstock_mod.fit(rstock_Xt, rstock_Yt, epochs = 10, batch_size = 200)
  rstock_prd = rstock_mod.predict(rstock_Xs)

  print("\nEvaluation of Rolls Royce on Stacked-LSTM\n")

  print("Score of R2   :", rstock_r2sc(rstock_Ys, rstock_prd)*100)
  print("Score of MAE  :", rstock_mae(rstock_Ys, rstock_prd))
  print("Score of MSE  :", rstock_mse(rstock_Ys, rstock_prd))
  print("Score of RMSE :", rstock_math.sqrt(rstock_mse(rstock_Ys, rstock_prd)))
```
