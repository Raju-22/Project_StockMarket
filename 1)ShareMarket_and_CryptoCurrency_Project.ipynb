```python
import pandas as gs_pd
```

```python
gs_data = gs_pd.read_csv('/content/drive/MyDrive/Datasets/BTC-USD.csv')
```

```python
gs_data
```

```python
gs_data.shape
```

```python
gs_data.info()
```

```python
print(gs_data.isnull().sum())
```

```python
gs_data = gs_data.dropna()
```

```python
# Check for duplicate rows in the entire DataFrame
duplicates = gs_data[gs_data.duplicated(keep=False)]

print("Duplicate rows in the DataFrame:", len(duplicates))
```

```python
gs_data.describe()
```

**Convert 'Date' column to datetime format**

```python
gs_data['Date'] = gs_pd.to_datetime(gs_data['Date'])
```

```python
gs_data
```

**Plotting the Closing Price:**

```python
import matplotlib.pyplot as gs_plt
```

```python
gs_plt.style.use('seaborn-v0_8-dark-palette')
gs_plt.figure(figsize=(14, 7))

gs_plt.figure(figsize=(14, 7))
gs_plt.plot(gs_data['Close'], label='Closing Price')
gs_plt.title('Bitcoin Closing Prices Over Time')
gs_plt.xlabel('Date')
gs_plt.ylabel('Closing Price')
gs_plt.legend()
gs_plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose as gs_sea_decomp
```

```python
decomposition = gs_sea_decomp(gs_data['Close'], model='multiplicative', period=365)
decomposition.plot()
gs_plt.show()
```

```python
gs_data.columns
```

```python
gs_data = gs_data[['Date','Close']]
```

```python
gs_data.shape
```

```python
gs_data.to_csv('/content/drive/MyDrive/Datasets/preprocessed_BIT_data.csv', index = False)
```
